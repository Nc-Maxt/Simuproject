```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML
from ust import Grafo, erase_loops

grafo = Grafo()
arr, d = grafo.random_walk(np.random.randint(0, 10, 2))
M = grafo.grid

m, n = M.shape
fig, ax = plt.subplots()
ax.set_xlim(-0.5, n-0.5)
ax.set_ylim(-0.5, m-0.5)
ax.set_xticks(np.arange(M.shape[0]))
ax.set_yticks(np.arange(M.shape[1]))
ax.grid(True)
xdata, ydata = [], []
(line,) = ax.plot([], [], color='blue', marker='o', linestyle='-', lw=2)

def init():
    line.set_data([], [])
    return line,

def update(frame):
    x, y = arr[frame]

    xdata.append(x)
    ydata.append(y)

    line.set_data(ydata, xdata)
    return line,


anim = FuncAnimation(fig,
                     update,
                     frames=len(arr),
                     init_func=init,
                     blit=True,
                     interval=500)

HTML(anim.to_jshtml())
```
